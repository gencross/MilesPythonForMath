```python
def collatz_verbose(n):
    while n != 1:
        print(n)
        if n % 2 == 0:
            n //= 2
        else:
            n = (3 * n + 1) // 2
    return
```

```python
def collatz_iterations(n):
    iterations = 0
    while n != 1:
        if n % 2 == 0:
            n //= 2
        else:
            n = (3 * n + 1) // 2
        iterations += 1
    return iterations
```

```python
def collatz_check(n):
    while n-1:
        if n % 2 == 0:
            n //= 2
        else:
            n = (3 * n + 1) // 2
    return True
```

```python
for i in range(2,10**8):
    if collatz_check(i) != True:
        print("Fields Medal!!!", i)
print("Check complete.")    
```

```python
def find_max_iter(n):
    max_iter = 0
    max_iter_n = 0
    for i in range(2,n+1):
        i_iter = collatz_iterations(i)
        if i_iter > max_iter:
            max_iter = i_iter
            max_iter_n = i
    return max_iter_n, max_iter
```

```python
find_max_iter(10**7)
```
